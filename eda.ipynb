```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d
```

<big>Doing some data visualizations!</big>

```python
cook_df = pd.read_csv('./samples/cook.csv')
flick_df = pd.read_csv('./samples/flickup.csv')

# cook_df.head()
#flick_df.head()

cook_x = cook_df["Ax"].to_numpy()
cook_y = cook_df["Ay"].to_numpy()
cook_z = cook_df["Az"].to_numpy()
cook_t = cook_df["Time"].to_numpy()

flick_x = flick_df["Ax"].to_numpy()
flick_y = flick_df["Ay"].to_numpy()
flick_z = flick_df["Az"].to_numpy()
flick_t = flick_df["Time"].to_numpy()
```

```python
#Cook Plots overlayed

fig, ax = plt.subplots()

ax.plot(cook_t, cook_x, label='X Acceleration')
ax.plot(cook_t, cook_y, label='Y Acceleration')
ax.plot(cook_t, cook_z, label='Z Acceleration')

ax.set_xlabel('Time')
ax.set_ylabel('Acceleration')
ax.legend()

plt.show()

```

```python
fig1, ax1 = plt.subplots()
ax1.plot(cook_t, cook_x)
ax1.set_xlabel('Time')
ax1.set_ylabel('X Acceleration')

fig2, ax2 = plt.subplots()
ax2.plot(cook_t, cook_y, color= 'orange')
ax2.set_xlabel('Time')
ax2.set_ylabel('Y Acceleration')

fig3, ax3 = plt.subplots()
ax3.plot(cook_t, cook_z, color = 'green')
ax3.set_xlabel('Time')
ax3.set_ylabel('Z Acceleration')

plt.show
```

```python
#Flick plots
fig, ax = plt.subplots()

ax.plot(flick_t, flick_x, label='X Acceleration')
ax.plot(flick_t, flick_y, label='Y Acceleration')
ax.plot(flick_t, flick_z, label='Z Acceleration')

ax.set_xlabel('Time')
ax.set_ylabel('Acceleration')
ax.legend()

plt.show()
```

```python
fig1, ax1 = plt.subplots()
ax1.plot(flick_t, flick_x)
ax1.set_xlabel('Time')
ax1.set_ylabel('X Acceleration')

fig2, ax2 = plt.subplots()
ax2.plot(flick_t, flick_y, color= 'orange')
ax2.set_xlabel('Time')
ax2.set_ylabel('Y Acceleration')

fig3, ax3 = plt.subplots()
ax3.plot(flick_t, flick_z, color = 'green')
ax3.set_xlabel('Time')
ax3.set_ylabel('Z Acceleration')

plt.show
```
